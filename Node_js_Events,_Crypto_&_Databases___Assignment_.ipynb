```python
# Question 1: Create a custom event emitter that fires a 'userLoggedIn' event with a username payload. Set up a listener to print a welcome message when the event occurs.

'''
// event-emitter.js
const EventEmitter = require('events');

const emitter = new EventEmitter();

// Listener
emitter.on('userLoggedIn', (username) => {
  console.log(`Welcome, ${username}! You have successfully logged in.`);
});

// Fire the event
emitter.emit('userLoggedIn', 'Alice');
emitter.emit('userLoggedIn', 'Bob');

//in terminal run it
node event-emitter.js

//output

Welcome, Alice! You have successfully logged in.
Welcome, Bob! You have successfully logged in.

'''
```

```python
# Question 2: Use the crypto module to create an SHA256 hash for the string "password123".

'''
// hash.js
const crypto = require('crypto');

const hash = crypto.createHash('sha256').update('password123').digest('hex');
console.log('SHA256 Hash:', hash);

//in terminal run it
node hash.js
'''
```

```python
# Question 3: Use the crypto module to generate a secure, random Version 4 UUID.

'''
// uuid.js
const crypto = require('crypto');

const uuid = crypto.randomUUID();
console.log('UUID:', uuid);

//in terminal run it
node uuid.js
'''
```

```python
# Question 5: Write the minimal code for an Express.js server that listens on port 3000 and responds to GET requests at the root URL (/) with "Hello, Express!".

'''
// server.js
const express = require('express');
const app = express();

app.get('/', (req, res) => {
  res.send('Hello, Express!');
});

app.listen(3000, () => {
  console.log('Server listening on http://localhost:3000');
});

//in terminal run it
npm install express
node server.js
'''
```

```python
# Question 6: Write a SQL statement to create a users table with an auto-incrementing primary key, an email, a password hash, and a creation timestamp.

'''
CREATE TABLE users (
  id         INT           NOT NULL AUTO_INCREMENT,
  email      VARCHAR(255)  NOT NULL UNIQUE,
  password   VARCHAR(255)  NOT NULL,
  created_at TIMESTAMP     NOT NULL DEFAULT CURRENT_TIMESTAMP,
  PRIMARY KEY (id)
);

DESCRIBE users;
'''
```

```python
# Question 8: Write a SQL query to select the id and email of the user whose email is "test@example.com".

'''
SELECT id, email
FROM users
WHERE email = 'test@example.com';
'''
```

```python
# Question 9: Write a SQL statement to update the email of the user with an id of 1 to "new.email@example.com".

'''
UPDATE users
SET    email = 'new.email@example.com'
WHERE  id = 1;

'''
```

```python
# Question 10: Given a users table and a posts table (with a user_id foreign key), write a query to retrieve all posts and the email of the user who wrote them.

'''
SELECT posts.id,
       posts.title,
       posts.content,
       posts.created_at,
       users.email AS author_email
FROM   posts
JOIN   users ON posts.user_id = users.id;

'''
```
